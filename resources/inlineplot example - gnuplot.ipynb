```elixir
Boyle.mk("inlineplot")
Boyle.activate("inlineplot")
Boyle.install({:gnuplot, "~> 1.19"})
```

```elixir
# Elixir Gnuplot discards the output, so we need to output to a file
Gnuplot.plot([
  ~w|set terminal pngcairo size 800,262 enhanced font 'Verdana,10'|a,
  ~w|set output 'sinx.png'|a,
  [:plot, 'sinh(x)', :title, "Sine Wave"]
  ])
  
# Wait for Gnuplot to write the image
:timer.sleep(1000)

{:"this is an inline image",src: "sinx.png"}
```
